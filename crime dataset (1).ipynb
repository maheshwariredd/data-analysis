Remove un named columns
----


```python
df_cleaned = df.rename(columns={
    "Sl.No": "Serial_No",
    "Heads of Crime": "Crime_Head",
    "Major Heads": "Major_Crime_Type",
    "Minor Heads": "Minor_Crime_Type",
    "During the current year upto the end of month under review": "Current_Year_Upto_Month",
    "During the corresponding month of previous year": "Prev_Year_Same_Month",
    "During the previous month": "Prev_Month",
    "During the current month": "Current_Month"
})
```

```python
df_cleaned
```

```python
df_cleaned = df_cleaned.drop(columns=["Unnamed: 8", "Unnamed: 9", "Unnamed: 10", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13"], errors='ignore')
```

```python
df_cleaned
```

```python
df_cleaned["Minor_Crime_Type"] = df_cleaned["Minor_Crime_Type"].fillna("Unknown")
```

```python
df_cleaned
```

```python
num_cols = ["Current_Year_Upto_Month", "Prev_Year_Same_Month", "Prev_Month", "Current_Month"]
```

```python
num_cols
```

```python
df_cleaned[num_cols] = df_cleaned[num_cols].fillna(0).astype(int)
```

```python
df_cleaned[num_cols]
```

```python
sns.set_style("whitegrid")
```

Top 10 Major Crime Types by Frequency
---

```python
top_crimes = df_cleaned["Major_Crime_Type"].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_crimes.values, y=top_crimes.index, palette="coolwarm")
plt.xlabel("Number of Cases")
plt.ylabel("Major Crime Type")
plt.title("Top 10 Major Crime Types by Frequency")
plt.show()
```

Histogram of Crimes in the Current Month
-

```python
plt.figure(figsize=(10,6))
sns.histplot(df_cleaned["Current_Month"], bins=20, kde=True, color='blue')
plt.xlabel("Number of Crimes")
plt.ylabel("Frequency")
plt.title("Histogram of Crimes in the Current Month")
plt.show()
```

Pie Chart of Major Crime Types
---

```python
plt.figure(figsize=(5,5))
df_cleaned["Major_Crime_Type"].value_counts().nlargest(5).plot(kind='pie', autopct='%1.1f%%', colormap='Set3')
plt.ylabel("Major Crime Type")
plt.title("Distribution of Top 5 Major Crime Types")
plt.show()
```

 Histogram of Crimes in the Previous Year
--


```python
plt.figure(figsize=(10, 5))
sns.histplot(df_cleaned["Prev_Year_Same_Month"], bins=20, kde=True, color='red')
plt.xlabel("Number of Crimes")
plt.ylabel("Frequency")
plt.title("Histogram of Crimes in the Previous Year")
plt.show()
```

Pie Chart of Major Crime Types in Previous Year
---

```python
plt.figure(figsize=(5,5))
df_cleaned.groupby("Major_Crime_Type")["Prev_Year_Same_Month"].sum().nlargest(5).plot(kind='pie', autopct='%1.1f%%', colormap='Pastel1')
plt.ylabel("Major Crime Type")
plt.title("Distribution of Top 5 Major Crime Types in Previous Year")
plt.show()
```

Bar Graph of Accidents
--


```python
accidents_data = df_cleaned[df_cleaned["Major_Crime_Type"].str.contains("Accident", case=False, na=False)]
plt.figure(figsize=(12, 6))
sns.barplot(x=accidents_data["Current_Month"], y=accidents_data["Minor_Crime_Type"], palette="Blues_r")
plt.xlabel("Number of Accidents")
plt.ylabel("Accident Type")
plt.title("Bar Graph of Different Types of Accidents")
plt.show()
```
