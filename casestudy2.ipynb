```python
df['S.NO'] = range(1, len(df) + 1)
df
```

```python
df['SECTION'] = df['SECTION'].fillna('SIGMA')
df['SECTION'] = df['SECTION'].replace('', 'SIGMA')
print("Updated DataFrame with missing sections replaced by 'SIGMA':")
df
```

```python
df=df.fillna(0)
df
```

```python
df.info()
```

```python
df['DV'] = pd.to_numeric(df['DV'], errors='coerce')
df['DV'] = df['DV'].fillna(0)
df['DV'] = df['DV'].astype(int)
```

```python
df
```

```python
df['M2'] = pd.to_numeric(df['M2'], errors='coerce')
df['M2'] = df['M2'].fillna(0)
df['M2'] = df['M2'].astype(int)
```

```python
df
```

```python
df['PP'] = pd.to_numeric(df['PP'], errors='coerce')
df['PP'] = df['PP'].fillna(0)
df['PP'] = df['PP'].astype(int)
```

```python
df
```

```python
df['BEEE'] = pd.to_numeric(df['BEEE'], errors='coerce')
df['BEEE'] = df['BEEE'].fillna(0)
df['BEEE'] = df['BEEE'].astype(int)
```

```python
df
```

```python
df['FL'] = pd.to_numeric(df['FL'], errors='coerce')
df['FL'] = df['FL'].fillna(0)
df['FL'] = df['FL'].astype(int)
```

```python
df
```

```python
df['FIMS'] = pd.to_numeric(df['FIMS'], errors='coerce')
df['FIMS'] = df['FIMS'].fillna(0)
df['FIMS'] = df['FIMS'].astype(int)
```

```python
df
```

calculate the total marks of each subject
-----

```python
df['Total']=df['DV']+df['M2']+df['PP']+df['BEEE']+df['FL']+df['FIMS']
df
```

add the percentage column to the table 
----

```python
df['Percentage']=round((df['Total']/120)*100,2)
df
```

giving grades based on the score
---

```python
df.loc[df["Total"]>110,["Grade"]]="A+"
df.loc[(df["Total"]>90) & (df["Total"]<=110),["Grade"]]="A"
df.loc[(df["Total"]>70) & (df["Total"]<=90),["Grade"]]="B+"
df.loc[(df["Total"]>50) & (df["Total"]<=70),["Grade"]]="B"
df.loc[(df["Total"]<=50),["Grade"]]="Fail"
df
```

failed students based on the marks
---

```python
failed_students_Total = df[df["Total"] < 50]
print("DataFrame of students who failed in subjects:")
print(failed_students_Total)
```
